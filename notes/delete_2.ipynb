```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("database_movies.csv")
```

```python
movies
```

```python
movies["genres"]=movies["genres"].replace('(no genres listed)','None')
```

```python
movies.drop(movies[movies['genres'] == "None"].index, inplace = True)
```

```python
movies
```

```python
import re
def clean_title(title):
    return re.sub("[^a-zA-Z ]", "", title)
movies["clean_title"] = movies["title"].apply(clean_title)
```

```python
movies
```

```python
def clean_genres(genres):
    return re.sub("[^a-zA-Z ]", " ", genres)
movies["clean_genres"] = movies["genres"].apply(clean_genres)
```

```python
movies
```

```python
movies = movies[["movieId", "clean_title", "clean_genres"]]
```

```python
movies
```

```python
movies['clean_title'] = [x.strip('( )').split(',') for x in movies['clean_title']]
```

```python
movies['clean_genres'] = [x.strip('( )').split(',') for x in movies['clean_genres']]
```
